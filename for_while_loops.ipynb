#for loop

A for loop is a programming construct that executes a block of code repeatedly until a condition is met. It's a control flow statement that specifies iteration, and is often used when the number of iterations is known before entering the loop

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
```

```python
# Let have for loop on list
country = ["India", "UK", "USA", "Germany"]  # 4... 1,2,3,4....  0,1,2,3

for ct in country:
    print(ct)
```

```python
# loop through string
for x in "banana":
  print(x)
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
  if x == "banana":
    break
```

```python
for x in range(6):
  print(x)
```

```python
for i in range(20,4000000): # range(start, stop)
    print(i)
```

```python
for i in range(0, 20, 5): # using range(start, stop, step)
    print(i)
```

```python
# range(): generating sequence of number 1, 2, 3, 4, 5
for i in range(5): # range(stop)
    print(i)
```

```python
country = ["India", "UK", "USA"] # 0,1,2,3

for i in range(len(country)):
  print(country[i])

# country[0]
# country[1]
# country[2]
# country[3]

print(len(country))
```

##The while Loop
With the while loop we can execute a set of statements as long as a condition is true

```python
count = 1

while count <= 5:   # while will run untill value of count is 5. If value is 6, it will not run
    print(count)    # printing value of count
    count += 1      # count = count + 1

#Check count <= 5..... count = 1 that mean condition true, condition holds

# 1st interation
# Step 1: print the value of count 1
# Step 2: count will be incremented by 1 count = count + 1

# Now updated Count = 2
# while count <= 5: still condition is true

# 2nd interation
# Step 1: print the value of count 2
# Step 2: count will be incremented by 1 count = count + 1 = 2 + 1 = 3

# Now updated Count = 3
# while count <= 5: still condition is true

# 3rd interation
# Step 1: print the value of count 3
# Step 2: count will be incremented by 1 count = count + 1 = 3 + 1 = 4

# Now updated Count = 4
# while count <= 5: still condition is true

# 4th interation
# Step 1: print the value of count 4
# Step 2: count will be incremented by 1 count = count + 1 = 4 + 1 = 5

# Now updated Count = 5
# while count <= 5: still condition is true

# 5th interation
# Step 1: print the value of count 5
# Step 2: count will be incremented by 1 count = count + 1 = 5 + 1 = 6

# Now updated Count = 6
# while count <= 5: Now condition false
```

```python
#Exit the loop when i is 3:

i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
#Continue to the next iteration if i is 3:

i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```
